```python
#import neccessary dependencies
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#count of items
result = df['itemDescription'].value_counts()

print(result)
```

```python
#transform data
df['transaction'] = df['Member_number'].astype('str')+''+df['Date'].astype('str')
df.head()
```

```python
#pivot the tables to convert items into columns
data= pd.crosstab(df['transaction'], df['itemDescription'])
data.head()
```

```python
#encoding all values using one hot encoding 
def encode(item_freq):
    res = 0
    if item_freq > 0:
        res = 1
     
    return res

basket_input = data.applymap(encode)
```

```python
basket_input.head()
```

```python
pip install mlxtend
```

```python
#using the Apriori algorithm from the mlxtend python package 
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

#set the rules for association mining
frequent_itemsets = apriori(basket_input, min_support=0.01, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="lift")
rules.head(10)

```

```python
rules.sort_values(["support", "confidence", "lift"], axis = 0, ascending = False).head(8)
```
